# Matplotlib Tutorial
## what is matplotlib?
- a plotting library for the Python programming language and its numerical mathematics extension NumPy
- it provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK

## Two ways to use matplotlib
1. pyplot - a MATLAB-like interface
   1. pyplot is a collection of command-style functions that make matplotlib work like MATLAB.
   2. pyplot is state-based, which means that it keeps track of the current figure and plot, and applies functions to the current figure or plot.
   3. pyplot is not suitable for creating complex plots or for embedding plots into applications.

2. object-oriented API
   1. object-oriented API is more powerful than pyplot
   2. object-oriented API is more flexible and more suitable for creating complex plots or for embedding plots into applications.

## Main components of matplotlib
1. Figure - a figure is a top-level container for all plot elements.
   
2. Axes - an axes is a container for a single plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 50)
y1 = 2*x + 1
y2 = x**2
# plt.figure()
plt.plot(x, y1)
# plt.show()
plt.plot(x, y2)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x, y1,color='blue',linewidth=3,linestyle='--',alpha=0.5)
plt.show()
plt.plot(x, y2,color='red',linewidth=3,linestyle='-.',alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x, y1,color='blue',linewidth=3,linestyle='--',label='line1')
# plt.show()
plt.plot(x, y2,color='red',linewidth=3,linestyle='-.',label='line2')
# plt.xlabel('x-axis')
# plt.ylabel('y-axis')
plt.title('my test plot') #loc: Literal["left", "center", "right"], default: "center"
plt.legend(loc='lower right') #check the loc strings that is available to use at: https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.legend.html#matplotlib.figure.Figure.legend
plt.xticks(())
plt.yticks(())
ax=plt.gca()
ax.spines['right'].set_color('None')
ax.spines['top'].set_color('None')
ax.spines['bottom'].set_color('None')
ax.spines['left'].set_color('None')
plt.show()
```

Y ticks and X ticks

```python
x=np.linspace(-3, 3, 50)
y1=2*x+1
y2=x**2
plt.figure(figsize=(8,5))
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim(0, 2)
plt.ylim(0, 3)
# set new sticks
new_ticks= np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
# set y ticks
plt.yticks([-2.1, -1.8, -1, 1.22, 3],
           ['really bad $\\alpha$', r'bad', r'normal', r'good', r'really good $\beta$'],rotation=45)

# get current axis
ax=plt.gca()
ax.spines['right'].set_color('None')
ax.spines['top'].set_color('None')
plt.title('title')
plt.show()
```

Annotation


```python
x=np.linspace(-3, 3, 50)
y=2*x+1
plt.figure(figsize=(8,5))
plt.plot(x, y, linewidth=1, color='black',linestyle='-',alpha=0.6)
# plt.scatter(x, y,s=9,color='blue',alpha=0.5)
ax=plt.gca()
ax.spines['right'].set_color('None')
ax.spines['top'].set_color('None')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
# annotation
x0=1
y0=2*x0+1
plt.scatter(x0, y0,s=30,color='red')
plt.text(x0+0.2, y0-0.3, 'A(x0,y0)', ha='left', va='bottom', fontsize=12, alpha=0.5)
# plt.annotate('A', xy=(x0, y0), xycoords='data', xytext=(+30, -30), textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

plt.show()
```

Bar plot

```python
n=5
X=np.arange(n)
Y1=(1-X/float(n)) * np.random.uniform(0.5, 1.0, n)
Y2=(1-X/float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')
for x,y in zip(X,Y1):
    plt.text(x+0.1, y-0.12, '%.2f' % y, ha='center', va= 'center')
#please do the same annotation for Y2?
plt.show()
```

User figure and axes object: plt.subplots 

```python
fig,axes=plt.subplots(2, 2, figsize=(8,6),sharex=True, sharey=True)
color=['red','black','blue','green']
colors=np.array(color).reshape(2,2)

for i in range(axes.shape[0]):
    for j in range(axes.shape[1]):
        axes[i,j].plot([1,2,3],[1,3,2],color=colors[i,j])
        axes[i,j].spines['right'].set_color('None')
        axes[i,j].spines['top'].set_color('None')
        if i==0:
            axes[i,j].spines['bottom'].set_color('None')            
            axes[i,j].xaxis.set_visible(False)
        if j==1:
            axes[i,j].spines['left'].set_color('None')
            axes[i,j].yaxis.set_visible(False)
        axes[i,j].set_title('subplot %s'%(i*2+j+1))

fig.tight_layout()
# fig.suptitle('Figure 1: Two subplots with shared axes')
plt.show()

```

```python
fig,axes=plt.subplots(figsize=(8,6))
np.random.seed(1)
x=np.arange(1,7)
y=np.random.randint(1,10,6)
axes.plot(x,y)
axes.set_xticks(x)
axes.set_yticks([1,3,5,7,9])
# axes.set_xticklabels(['a','b','c','d','e','f'])
axes.set_yticklabels(['1','3','5','7','9'])

axes1=fig.add_subplot(*[5,5,17])
axes1.plot(x,y)
plt.show()
```

```python
x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2
fig,ax=plt.subplots()
ax1=ax.twinx()
ax.plot(x,y1,color='red',label='linear')
ax1.plot(x,y2,color='green',label='square')
ax.set_xlabel('x')
ax.set_ylabel("y1",color="green",rotation=0)
ax1.set_ylabel("y2",color="red",rotation=0)
ax1.legend(loc='upper left')
ax.legend(loc='lower left')
plt.show()
```
